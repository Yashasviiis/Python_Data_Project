```python
import numpy as np
```

```python
salary_array = np.array(salary)
salary_array
```

```python
sum_salary = np.sum(salary_array)
print(sum_salary)
```

```python
med_salary = np.median(salary_array)
print(med_salary)
```

```python
cumsum_salary = np.cumsum(salary_array)
cumsum_salary
```
